```python
df= pd.read_csv("CountryTable.csv",index_col=2)
df
```

```python
#Filtro de Filas vs Columnas
filtro4=df.loc[["North America","Europe"],["population","life_expectancy"]]
filtro4
#Convertir archivo filtrado a CSV
filtro4.to_csv("Entregable1Filtro4.csv")
#Descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable1Filtro4.csv")
```

```python
#Filtro por cabecera
filtro5=df.head(8)
filtro5
#Convertir archivo filtrado a CSV
filtro5.to_csv("Entregable1Filtro5.csv")
#Descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable1Filtro5.csv")
```

```python
#Filtro por comparaciÃ³n
filtro6=df[df["population"] > 10000]
filtro6
#Convertir archivo filtrado a CSV
filtro6.to_csv("Entregable1Filtro6.csv")
#Descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable1Filtro6.csv")
```

```python
#Filtro &
filtro7=df[(df["life_expectancy"] > 60) & (df["region"] == "Southern Europe")]
filtro7
#Convertir archivo filtrado a CSV
filtro7.to_csv("Entregable1Filtro7.csv")
#Descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable1Filtro7.csv")
```

```python
#Filtro or
filtro8=df[(df["life_expectancy"] > 30) | (df["region"] == "Caribbean")]
filtro8
#Convertir archivo filtrado a CSV
filtro8.to_csv("Entregable1Filtro8.csv")
#Descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable1Filtro8.csv")
```

```python
#Filtro not
filtro9=df[~(df["population"] >= 50000) & ~(df["region"] == "Caribbean")]
filtro9
#Convertir archivo filtrado a CSV
filtro9.to_csv("Entregable1Filtro9.csv")
#Descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable1Filtro9.csv")
```
